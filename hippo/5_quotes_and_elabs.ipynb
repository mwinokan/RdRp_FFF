## Assign subsites to scaffolds

```python
animal.db.insert_subsite(1, "Cavity B")
animal.db.insert_subsite(1, "Green Site/RNA tunnel")
animal.db.insert_subsite(1, "Primer grip/N-pocket")
animal.db.insert_subsite(1, "Thumb Site 2")
```

```python
lookup = {
    '[Other] Cavity B (S3) (P2)': 1,
    '[Other] Green Site/RNA tunnel (P2)': 2,
    '[Other] Primer grip/N-pocket (P2)': 3,
    '[Other] Thumb Site 2 (P2)': 4,
}
```

```python
for tag, site in lookup.items():
    poses = animal.poses(tag=tag)

    for pose in mrich.track(poses):
        animal.db.insert_subsite_tag(pose_id=pose.id, subsite_id=site, name=None, commit=False)

animal.db.commit()
```
