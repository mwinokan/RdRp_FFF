## Parse Jasmin Excel

```python
csvs = {
    "MW_first14_retro_batch1.csv":"MW",
    "MW_top4_sucos_batch1.csv":"MW",
    "MW_top_sucos_retro_batch2.csv":"MW",
    "Retro_OK_batch1-2024-12-10-A_JA_cavityB.csv":"JA",
    "Retro_OK_batch1-2024-12-10-A_JA_rest.csv":"JA",
    "Retro_OK_batch2-2024-12-12-A_JA.csv":"JA",
}
```

```python
CURATION = {}
for csv,name in csvs.items():
    mrich.h3(csv)

    path = Path('../curation')/csv

    print(path)
    
    df = pd.read_csv(path, skiprows=1)

    for i,row in df.iterrows():
        pose_id = row["HIPPO Pose ID"]
        pose = animal.poses[pose_id]
        curation = []
        for col in row.to_dict():
            if col.endswith("-text"):
                value = row[col]
                # print(col, value, curation)
                if not pd.isna(value):
                    curation.append(f"{name}: {value}")
        
        # mrich.var(pose.id, ",".join(curation))

        if not curation:
            continue
        
        if pose.id not in curation:
            CURATION[pose.id] = []
            
        CURATION[pose.id] += (curation)
    
    # break
```
