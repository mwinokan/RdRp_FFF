```python
ISETS = {}
```

```python
def interaction_overlap(poses):
    for pose in poses:
        if pose.id in ISETS:
            continue
        ISETS[pose.id] = set(pose.interactions.type_residue_number_chain_triples)

    ids = poses.ids
    
    s = 0
    for pose_j in ids:
        iset_j = ISETS[pose_j]
        for pose_k in ids:
            
            iset_k = ISETS[pose_k]
            
            intersection = iset_j & iset_k
            diff1 = iset_j - iset_k
            diff2 = iset_k - iset_j
            
            if intersection and diff1 and diff2:
                s += 1
    
    print("overlap", poses, s)
    return s
```

```python
TAGMAP = {
    '[Other] Cavity B (S3) (P2)':"Cavity B merge",
    '[Other] Green Site/RNA tunnel (P2)':"Green Site/RNA tunnel merge",
    '[Other] Thumb Site 2 (P2)':"Thumb Site 2 merge",
    '[Other] Primer grip/N-pocket (P2)':"Primer grip/N-pocket merge",
}
```

```python
data = []
for hit_tag, merge_tag in TAGMAP.items():
    print(hit_tag, merge_tag)
    hits = animal.poses(tag=hit_tag)
    merges = animal.poses(tag=merge_tag)
    
    acceptable = merges.get_by_tag(tag="acceptable")
    
    knitwork_pure = merges.get_by_tag(tag="Knitwork pure merge")
    knitwork_impure = merges.get_by_tag(tag="Knitwork impure merge")
    fragmenstein_pure = merges.get_by_tag(tag="Fragmenstein pure merge")
    
    acceptable_knitwork_pure = knitwork_pure.get_by_tag(tag="acceptable")
    acceptable_knitwork_impure = knitwork_impure.get_by_tag(tag="acceptable")
    acceptable_fragmenstein_pure = fragmenstein_pure.get_by_tag(tag="acceptable")
    
    data.append({
        "hypothesis":hit_tag,
        "hits":hits,
        "overlap opportunities":hits.interaction_overlap_score,
        "all merges":merges,
        "acceptable merges":acceptable,
        "all knitwork pure": knitwork_pure,
        "all knitwork impure": knitwork_impure,
        "all fragmenstein pure": fragmenstein_pure,
        "acceptable knitwork pure": acceptable_knitwork_pure,
        "acceptable knitwork impure": acceptable_knitwork_impure,
        "acceptable fragmenstein pure": acceptable_fragmenstein_pure,
    })
df = pd.DataFrame(data)
df
```

```python
count_data = []
for i,row in df.iterrows():
    d = {}
    for col, value in row.to_dict().items():
        # value = row[col]
        if not isinstance(value, int) and not isinstance(value, str):
            value = len(value)
        d[col] = value
    count_data.append(d)
count_df = pd.DataFrame(count_data)
count_df
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(name="overlap opportunities", x=count_df["hypothesis"].values, y=count_df["overlap opportunities"].values))
fig.add_trace(go.Bar(name="acceptable merges", x=count_df["hypothesis"].values, y=count_df["acceptable merges"].values))
fig.update_layout(title="RdRp: overlap opportunities vs acceptable merges, by site", xaxis_title="hypothesis", yaxis_title="count")
```
