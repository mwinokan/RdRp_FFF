```python
# chosen_ids = animal.db.select_id_where(table='pose', key=f'pose_id IN {product_poses.str_ids} AND pose_metadata LIKE "%_fragalysis.sdf%"', multiple=True)
chosen_ids = animal.db.select_id_where(table='pose', key=f'pose_id IN {product_poses.str_ids} AND pose_metadata LIKE "%RdRp_knitwork_green_site_rna_tunnel_pure_fragalysis.sdf%"', multiple=True)
chosen_ids = [i for i, in chosen_ids]
```

```python
filtered_poses = animal.poses[chosen_ids]
```

```python
product_poses.interactive()
```

```python
for pose in mrich.track(filtered_poses):
    result = pose.score_inspiration(return_all=False, debug=False)
    # break
    # print(result)
    pose.metadata["Multi-SuCOS"] = result
    # break
```

```python
for pose in mrich.track(filtered_poses):
    path = Path([p for p in pose.metadata["exports"] if p.endswith("_fragalysis.sdf")][0])
    mergeset = path.name.split("_split")[0]
    pose.tags.add(f"{mergeset} Pose OK")
    pose.metadata["merge_set"] = mergeset
    # break
```

```python
df = filtered_poses.to_fragalysis(
    "Retro_OK_batch2.sdf",
    method="Retro_OK_batch2",
    submitter_institution="DLS", 
    submitter_name="Max Winokan", 
    submitter_email="max.winokan@diamond.ac.uk", 
    ref_url="https://github.com/mwinokan/RdRp_FFF"
)
```
